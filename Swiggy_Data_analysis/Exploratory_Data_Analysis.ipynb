```python
df_country=pd.read_excel("Country-Code.xlsx")
df_country.head(4)
```

```python
#i have joined two datasets by the help of merge function
final_df=pd.merge(df,df_country,on="Country Code",how="left")
final_df.head(3)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_values=final_df.Country.value_counts().values
country_values
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
#pie chart top three countries uses swiggy
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.2f%%")
```

Observation= Swiggy maximum transactions occured from india than US and Uk

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

Observations
1. when ratings are between 4.5-4.9 it is ->Excellent
2. when ratings are between 4.0-4.4 it is ->Very Good
3. when ratings are between 3.5-3.9 it is ->Good
4. when ratings are between 2.5-3.4 it is ->Average
5. when ratings are between 1.8-2.4 it is ->Poor
6. when rating is  0.0 it is ->Not rated

```python
ratings.head(3)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
plt.show()
```
